```python
# 1. Dataset Overview
print("Dataset Shape:", df.shape)
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
```

```python
# 2. Replace -99 with NaN for meaningful analysis
df.replace(-99, pd.NA, inplace=True)
```
